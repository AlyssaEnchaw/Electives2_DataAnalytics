```python
# The graph will only show the How confident in achieving your New Year's Resolution this year column since it is the only int data types, the rest is object and string
sns.distplot(df["How confident are you in achieving your New Year's resolution this year?"], bins = 10)
plt.title("Confidence of the interviewee on achieving New Year's Resolution")
plt.show()

print("\n")
#Age Group
ageGroup = df.groupby('Age Group').size()
ageGroup.plot.pie(autopct='%1.5f%%')
plt.title("Percentage of Age Group")
plt.legend(['17-20 years old','24-27 years old','21-23 years old'])
plt.show()

print("\n")
#Gender 
ageGroup = df.groupby('Gender').size()
ageGroup.plot.pie(autopct='%1.5f%%')
plt.title("Percentage of Gender")
plt.legend(['Other','Female','Gay','Male','Prefer not to say'])
plt.show()

print("\n")
#Religion
ageGroup = df.groupby('Religion').size()
ageGroup.plot.pie(autopct='%1.5f%%')
plt.title("Percentage of Religion")
plt.legend(['Catholic','Christian','None','Prefer not to say'])
plt.show()

print("\n")
#Professional Status
ageGroup = df.groupby('Please specify your current professional status').size()
ageGroup.plot.pie(autopct='%1.5f%%')
plt.title("Percentage of Professional Status")
plt.legend(['Full-time Employee','Student','Unemployed','Working Student'])
plt.show()

print("\n")
#Marital Status Status
ageGroup = df.groupby('What is your marital status?').size()
ageGroup.plot.pie(autopct='%1.5f%%')
plt.title("Percentage of Professional Status")
plt.legend(['Married','Prefer not to answer','Single'])
plt.show()

print("\n")
#Plan
ageGroup = df.groupby("Did you plan a New Year's resolution this 2023?").size()
ageGroup.plot.pie(autopct='%1.5f%%')
plt.title("Plan")
plt.legend(['No','Yes'])
plt.show()

print("\n")
#Achieve
ageGroup = df.groupby("Have you ever achieved a New Year's resolution in the past?").size()
ageGroup.plot.pie(autopct='%1.5f%%')
plt.title("Achieve?")
plt.legend(['No','Yes'])
plt.show()

print("\n")
#Reason
ageGroup = df.groupby("If you have made New Year's resolutions in the past that you did not achieve, what do you think was the reason?").size()
ageGroup.plot.pie(autopct='%1.5f%%')
plt.title("Reason?")
plt.show()

print("\n")
#Overcome
ageGroup = df.groupby("How do you plan to overcome any obstacles you may face in achieving your New Year's resolution?").size()
ageGroup.plot.pie(autopct='%1.5f%%')
plt.title("Overcome?")
plt.show()

```
